```python
x, sr = librosa.load(r"metal.00000.wav")
print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
def create_spectrogram(file):
    x, sr = librosa.load(file)
    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    # plt.colorbar()

    plt.axis('off')

    # Removing the .wav at the end of file
    split = file.split('.')
    filename = split[0]+split[1]

    #save
    plt.savefig(filename+'.png', bbox_inches = 'tight', pad_inches = 0)
    plt.close()
```

```python
create_spectrogram(r"metal.00000.wav")
```

Function to save a spectrogram to my Directory!

```python
def create_spectrogram_axis(file):
    x, sr = librosa.load(file)
    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()

    # plt.axis('off')

    # Removing the .wav at the end of file
    split = file.split('.')
    filename = split[0]+split[1]

    #save
    plt.savefig(filename + '_axis' + '.png')
    plt.close()
```

```python
create_spectrogram_axis(r"WavFiles/rock.00000.wav")
```
